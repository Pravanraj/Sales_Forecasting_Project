```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from prophet import Prophet
```

```python
data = pd.read_csv('/Users/dhinakar/Downloads/College/Project/Akaike/data/Challenge 1 Data.csv', encoding='ISO-8859-1')
data.head()
```

```python
data.isnull().sum()
```

```python
# Drop rows  missing important values
data = data.dropna(subset=['InvoiceNo', 'Quantity', 'UnitPrice', 'InvoiceDate'])
```

```python
data=data[(data['Quantity'] > 0) & (data['UnitPrice'] >0)]#removes the cancels or return
data['InvoiceDate']=pd.to_datetime(data['InvoiceDate'])#invoice to date annd time
data['Revenue']=data['Quantity']*data['UnitPrice']#revenue=qty*unitprice
data['Revenue'].head()
```

```python
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
daily_sale = data.groupby(data['InvoiceDate'].dt.date)['Revenue'].sum().reset_index()#revenue of each date will be summed and indexed from first
daily_sale.columns=['ds','y']#datestamp and revenue
daily_sale.head()
```

```python
plt.figure(figsize=(14,5))
plt.plot(daily_sale['ds'], daily_sale['y'])
plt.title('Sales Trend')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.show()#Plot of sales trend
```

```python
from prophet import Prophet
model=Prophet(daily_seasonality=True,yearly_seasonality=True,weekly_seasonality=True)
model.fit(daily_sale)#fitting the prophet model
```

```python
#dataframe for the next 7 days
next_7_d = model.make_future_dataframe(periods=7)
forecast = model.predict(next_7_d)
```

```python
figure = model.plot(forecast)
plt.title('Forecast for next 7 days')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.tight_layout()
plt.show()
```

```python
daily_sales['ds'] = pd.to_datetime(daily_sales['ds'])
forecast['ds'] = pd.to_datetime(forecast['ds'])
comparison = pd.merge(daily_sales, forecast[['ds', 'yhat']], on='ds', how='left')
```

```python
plt.figure(figsize=(12, 6))
plt.plot(comparison['ds'], comparison['y'], label='Actual Sales', color='blue')
plt.plot(comparison['ds'], comparison['yhat'], label='Forecasted Sales', color='orange')
plt.title('Actual vs Forecasted Daily Sales')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.show()
```
